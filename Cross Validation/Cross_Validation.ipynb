**Importing the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

**Importing Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

**Data Collection & Preprocessing**

```python
# Loading Data into Pandas Dataframe
heart_disease_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
# Printing 5 rows of Dataset:
heart_disease_data.head()
```

```python
# Shape of Dataset:
heart_disease_data.shape
```

```python
# Check for Null Values:
heart_disease_data.isnull().sum()
```

```python
# Statistical Measures of Dataset:
heart_disease_data.describe()
```

```python
# Distribution of Target Variable:
heart_disease_data['target'].value_counts()
```

```python
# 1 --> Defective Heart
# 2 --> Healthy  Heart
```

**Splitting Features and Targets**

```python
X = heart_disease_data.drop(columns= 'target', axis=1)
Y = heart_disease_data['target']
```

```python
print(X)
```

```python
print(Y)
```

#**Train Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

**Comparing the Performance of Differernt Models**

```python
# List of Models:
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test():
  for model in models:
    # Training the model:
    model.fit(X_train, Y_train)
    # Evaluating the model:
    test_data_prediction = model.predict(X_test)
    accuracy = accuracy_score(Y_test, test_data_prediction)
    print("Accuracy Score of the Model", model, ' = ', accuracy)
```

```python
compare_models_train_test()
```

#**Cross Validation**

```python
cv_score_LR = cross_val_score(LogisticRegression(max_iter=1000),X,Y, cv=5)
print(cv_score_LR)

mean_accuracy_LR = sum(cv_score_LR)/len(cv_score_LR)
mean_accuracy_LR = mean_accuracy_LR*100
mean_accuracy_LR = round(mean_accuracy_LR,2)
print(mean_accuracy_LR)
```

**Support Vector Classifier**

```python
cv_score_SVC = cross_val_score(SVC(kernel='linear'),X,Y, cv=5)
print(cv_score_SVC)
mean_accuracy_SVC = sum(cv_score_SVC)/len(cv_score_SVC)
mean_accuracy_SVC = mean_accuracy_SVC*100
mean_accuracy_SVC = round(mean_accuracy_SVC,2)
print(mean_accuracy_SVC)
```

**RandomForestClassifier**

```python
cv_score_RFC = cross_val_score(RandomForestClassifier(),X,Y, cv=5)
print(cv_score_RFC)
mean_accuracy_RFC = sum(cv_score_RFC)/len(cv_score_RFC)
mean_accuracy_RFC = mean_accuracy_RFC*100
mean_accuracy_RFC = round(mean_accuracy_RFC,2)
print(mean_accuracy_RFC)
```

**Creating a Function to compare the models:**

```python
# List of Models:
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_cross_validation():

  for model in models:

    cv_score = cross_val_score(model, X,Y, cv=5)
    mean_accuracy = sum(cv_score)/len(cv_score)
    mean_accuracy = mean_accuracy*100
    mean_accuracy = round(mean_accuracy,2)
    print("Cross Validation Accuracies for ", model, '=', cv_score)
    print('Accuracy % of the ', model, mean_accuracy)
    print('-----------------------------------------------')
```

```python
compare_models_cross_validation()
```
